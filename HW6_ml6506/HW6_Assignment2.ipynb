```python
#  extra credit: make the function python 2 and 3 compatible so that it works on the PUI2016_Python3 kernel
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError
    
import ast
import pandas as pd
```

```python
def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urlencode({'format': format, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep=',')
    except ValueError as v:
        print (str(v))
```

```python
query = '''SELECT * FROM fb55.citibike; '''
table = get_data(query)
table.head(5)
```

```python
# Sort data by start_station_id, tripduration
# Only checking trips with duration <= 3 hours
# Only show the top/last 10 records (aka head and tail in SQL)

# sort by start_station_id
query = '''SELECT * FROM fb55.citibike
WHERE tripduration <= 3 * 360
ORDER BY start_station_id DESC
limit 10
'''
table = get_data(query)
table.head(10)
```

```python
# sort by tripduration
query = '''
SELECT * FROM fb55.citibike
WHERE tripduration <= 3 * 360
ORDER by tripduration DESC
LIMIT 10;
'''
table = get_data(query)
table.head(10)
```

```python
#sort by start_station_id and tripduration
query = '''
SELECT * FROM fb55.citibike
ORDER BY start_station_id ASC, tripduration DESC
LIMIT 10;
'''
table = get_data(query)
table.head(10)
```

```python
#only show last 10
query = '''SELECT * FROM fb55.citibike
ORDER BY start_station_id ASC, tripduration DESC
LIMIT 10;
'''
table = get_data(query)
table.tail(10)
```

```python
#List all unique start_station_id values
dist_query = '''SELECT DISTINCT start_station_id FROM fb55.citibike
                WHERE tripduration <= 3 * 3600
                ORDER BY start_station_id; '''
dist_table = get_data(dist_query)
dist_table.head()
```

```python
#  Aggregation functions:
# Count the number of trips (aka wc -l in SQL)
# Find the average/min/max trip duration

query = '''
SELECT count(start_station_id) as trip_count,
AVG(tripduration),
MIN(tripduration),
MAX(tripduration)
FROM fb55.citibike
'''
table = get_data(query)
table.head()
```

# Task 2  Working with date/time

```python
# Selecting trips started on Feb-02-2015 only
query = '''SELECT FROM fb55.citibike
WHERE starttime >= '2015-02-02 00:00' and starttime < '2015-02-03 00:00'
'''
tabel = get_data(query)
table.head(5)
```

```python
# Selecting trips started on the weekends
# What are average trip duration during weekends?
query = '''
SELECT avg(tripduration) as weekend_avg FROM fb55.citibike
WHERE extract(dow from starttime) in (0,6)
'''
table = get_data(query)
table.head()
```

```python
# Can we do the same for weekday?
query = '''
SELECT avg(tripduration) as weekdat_avg FROM fb55.citibike
WHERE extract(dow from starttime) in (1,2,3,4,5)
'''
table = get_data(query)
table.head()
```

# Task 3  Working with Space

```python
# Showing the list of start station locations(Using GROUP BY)
loc_query = '''SELECT start_station_name FROM fb55.citibike
                GROUP BY start_station_name; '''
loc_table = get_data(loc_query)
loc_table.head()
```

```python
# Showing the number of trips started per station
query = '''
SELECT count(start_station_id),
start_station_name
FROM fb55.citibike
GROUP BY start_station_id, start_station_name
'''
table = get_data(query)
table.head()
```

```python
#  but only for stations within 500m of Time Square!
# The coordinates of Time Square is (40.7577,-73.9857)
query = '''SELECT start_station_name, COUNT(bikeid) as trip_number FROM fb55.citibike
                WHERE (6371000 * acos(cos(radians(40.7577)) * cos(radians(start_station_latitude)) * cos(radians(start_station_longitude) - radians(-73.9857)) + sin(radians(40.7577)) * sin(radians(start_station_latitude)))) <= 500
                GROUP BY start_station_name; '''
table = get_data(query)
table
```

# Task 4  Putting it all together

```python
# Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare
query = '''SELECT start_station_name, AVG(tripduration) as weekends_avg_tripduration FROM fb55.citibike
                WHERE EXTRACT(DOW FROM starttime::date) in (0, 6) AND (6371000 * acos(cos(radians(40.7577)) * cos(radians(start_station_latitude)) * cos(radians(start_station_longitude) - radians(-73.9857)) + sin(radians(40.7577)) * sin(radians(start_station_latitude)))) <= 500
                GROUP BY start_station_name
                ORDER BY weekends_avg_tripduration DESC
                LIMIT 1; '''
table = get_data(query)
table
```
